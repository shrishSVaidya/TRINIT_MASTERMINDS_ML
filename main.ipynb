# Model Training


```python
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.metrics import accuracy_score
```

```python
cols = df1.columns
cols
```

```python
df1.isnull().sum()
```

```python
new_label = preprocessing.LabelEncoder()
X = df1.drop('label', axis=1)
y = df1['label']
```

```python
new_label.fit(y)
mapping = dict(zip(new_label.classes_, new_label.transform(new_label.classes_)))
y = new_label.fit_transform(y)
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 32)
```

```python
mapping
```

## SVC:

```python
from sklearn import svm
svc = svm.SVC()
```

```python
svc.fit(x_train, y_train)
```

```python
y_predict = svc.predict(x_test)
accuracy_score(y_test, y_predict)
```

## Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier(random_state=32, max_depth=10)
dtc.fit(x_train, y_train)
```

```python
y_pred = dtc.predict(x_test)
accuracy_score(y_test, y_pred)
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(max_depth=5, random_state=0, n_estimators = 100)
```

```python
rfc.fit(x_train, y_train)
```

```python
y_pred = rfc.predict(x_test)
accuracy_score(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```
